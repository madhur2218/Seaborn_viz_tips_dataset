# Distribution Plots

```python
import seaborn as sns 
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

```python
sns.distplot(df["tip"]) #histogram
```

```python
sns.distplot(df["total_bill"],bins=30,color='Red',vertical=True,kde=False) #histogram
```

```python
sns.distplot(df['size'], kde=True, bins=5,color='Green')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex',color='Black')
```

```python
df.head()
```

sns.jointplot(data=df,x='tip',y='size')

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```

```python
sns.pairplot(df) #Pairplot gives make various plots where data is having multiple numerical columns 
```

```python
sns.pairplot(df,hue='smoker',palette='Spectral')
```

```python
sns.pairplot(df,hue='day')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.kdeplot(df['total_bill']) # kde plot is used to check if my data set is normally distributed or not
```

```python
sns.kdeplot(df['tip'])
sns.rugplot(df['tip'])
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df, palette="coolwarm")
```

```python
sns.violinplot(data=df,x='time',y='tip',hue='sex')
```
